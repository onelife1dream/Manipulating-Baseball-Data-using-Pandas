#### 10)	Top 5 highest OBP (on base percentage) with at least 500 PAs in 1977?  (Dataframe)

PA = AB+BB+IBB+SH+SF
AVERAGE=H/AB
OBP = (H+BB+IBB+SH+SF)/AB

```python
df = pd.read_csv('Batting.csv')
```

```python
temp = df[df['yearID'] == 1977]
```

```python
temp = temp[temp['AB'] != 0]
```

```python
temp = temp.fillna(0)
```

```python
temp['PA'] = temp['AB']+temp['BB']+temp['IBB']+temp['SH']+temp['SF']
```

```python
temp = temp[temp['PA'] >= 500]
```

```python
temp['OBP'] = (temp['H']+temp['BB']+temp['IBB']+temp['SH']+temp['SF'])/temp['AB']
temp.sort_values('OBP', ascending = False)[:5]
```

#### 13)	Who led MLB(entire data set) with the most hits the most times?  And how many times?  (Dataframe, Number)

```python
df = pd.read_csv('Batting.csv')
```

```python
temp = df.groupby('yearID')['H'].max()
```

```python
years = df['yearID'].unique()
```

```python
name = []
```

```python
for i in years:
    name.append(df[(df['yearID']== i) & (df['H'] == temp[i])][['playerID','nameFirst','nameLast']])
result = pd.concat(name)
gresult = result.groupby(['nameFirst','nameLast']).count()
mosthit = gresult['playerID'].max()
gresult[gresult['playerID'] == mosthit]
```
